```python
# ================= CELL: ANALISIS LEARNING CURVE =================
import pandas as pd
import matplotlib.pyplot as plt
import os

# GANTI PATH INI dengan lokasi file CSV evaluasi kamu yang baru
CSV_LOG_PATH = r"D:\Media\Kuliah\Skripsi\ScripTI\Versi 4\Model\Final Model\finetuned_nusabert_base_extended\eval\similarity_evaluation_sts-val-extended_results.csv"

# Cek file
if not os.path.exists(CSV_LOG_PATH):
    print("‚ùå File CSV tidak ditemukan. Cek path-nya lagi.")
else:
    # Baca Data
    df = pd.read_csv(CSV_LOG_PATH)
    
    # Tampilkan Data Mentah
    print("Data Evaluasi per Epoch:")
    print(df[['epoch', 'steps', 'cosine_pearson', 'cosine_spearman']])

    # Cari Skor Terbaik
    best_epoch_idx = df['cosine_spearman'].idxmax()
    best_score = df.loc[best_epoch_idx, 'cosine_spearman']
    best_epoch = df.loc[best_epoch_idx, 'epoch']

    print("\n" + "="*40)
    print(f"üèÜ EPOCH TERBAIK: Epoch {best_epoch}")
    print(f"üìà Skor Spearman Tertinggi: {best_score:.4f}")
    print("="*40)

    # --- PLOT GRAFIK (Bahan Bab 4) ---
    plt.figure(figsize=(10, 6))
    
    # Plot Spearman
    plt.plot(df['epoch'], df['cosine_spearman'], marker='o', linestyle='-', linewidth=2, color='blue', label='Spearman Correlation')
    
    # Tandai titik terbaik
    plt.scatter(best_epoch, best_score, color='red', s=100, zorder=5, label=f'Best Epoch ({best_epoch})')
    plt.annotate(f'{best_score:.4f}', (best_epoch, best_score), textcoords="offset points", xytext=(0,10), ha='center')

    plt.title('Kurva Pembelajaran Model (Evaluasi pada Validation Set)', fontsize=14)
    plt.xlabel('Epoch (Tambahan)', fontsize=12)
    plt.ylabel('Skor Korelasi', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    # Simpan Gambar untuk Skripsi
    plt.savefig("grafik_learning_curve.png", dpi=300)
    plt.show()
    print("Grafik disimpan sebagai 'grafik_learning_curve.png'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob

# --- KONFIGURASI PATH ---
# 1. Path hasil training pertama (Epoch 1-4)
PATH_PHASE_1 = r"D:\Media\Kuliah\Skripsi\ScripTI\Versi 4\Model\Trained_SBERT\finetuned_all-nusabert-base_v1\eval"

# 2. Path hasil training tambahan (Epoch 5-10)
PATH_PHASE_2 = r"D:\Media\Kuliah\Skripsi\ScripTI\Versi 4\Model\Final Model\finetuned_nusabert_base_extended\eval"

# Fungsi untuk mencari file CSV evaluasi
def load_eval_log(folder_path, start_epoch_offset=0):
    # Cari file csv di folder eval
    csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
    if not csv_files:
        print(f"‚ùå Tidak ada CSV di: {folder_path}")
        return None
    
    # Ambil file pertama (biasanya cuma ada satu)
    df = pd.read_csv(csv_files[0])
    
    # Koreksi nomor Epoch
    # SBERT mereset epoch jadi 0 setiap start baru. Kita harus geser manual.
    # Namun, dari outputmu terlihat epoch dimulai dari 1.0.
    # Kita geser epoch Phase 2 agar nyambung (Misal Phase 1 selesai di epoch 4, Phase 2 mulai jadi epoch 5)
    df['global_epoch'] = df['epoch'] + start_epoch_offset
    
    return df

# --- PROSES PENGGABUNGAN ---
print("Memuat data evaluasi...")

# Load Phase 1 (Epoch 1-4) -> Offset 0
df1 = load_eval_log(PATH_PHASE_1, start_epoch_offset=0)

# Load Phase 2 (Epoch 1-6) -> Offset 4 (Karena melanjutkan setelah epoch 4)
# Catatan: Di outputmu epoch tertulis 1.0, 2.0.
# Jika Epoch 1.0 itu adalah lanjutan langsung, maka itu adalah Global Epoch 5.
df2 = load_eval_log(PATH_PHASE_2, start_epoch_offset=4)

if df1 is not None and df2 is not None:
    # Gabung
    df_full = pd.concat([df1, df2]).reset_index(drop=True)
    
    # Pastikan urut
    df_full = df_full.sort_values(by='global_epoch')
    
    print("\n--- DATA LENGKAP (EPOCH 1 - 10) ---")
    print(df_full[['global_epoch', 'cosine_spearman', 'cosine_pearson']])
    
    # Cari Puncak Tertinggi Sejagat Raya
    best_idx = df_full['cosine_spearman'].idxmax()
    best_epoch = df_full.loc[best_idx, 'global_epoch']
    best_score = df_full.loc[best_idx, 'cosine_spearman']
    
    print("\n" + "="*40)
    print(f"üèÜ GLOBAL BEST MODEL: Epoch {int(best_epoch)}")
    print(f"üìà Skor Spearman: {best_score:.4f}")
    print("="*40)
    
    # --- VISUALISASI ---
    plt.figure(figsize=(10, 6))
    
    # Garis Spearman
    plt.plot(df_full['global_epoch'], df_full['cosine_spearman'], 
             marker='o', linestyle='-', linewidth=2, color='green', label='Spearman Correlation')
    
    # Tandai Puncak
    plt.scatter(best_epoch, best_score, color='red', s=150, zorder=5, label=f'Best (Epoch {int(best_epoch)})')
    
    # Anotasi
    plt.annotate(f'{best_score:.4f}', (best_epoch, best_score), 
                 textcoords="offset points", xytext=(0,15), ha='center',
                 bbox=dict(boxstyle="round,pad=0.3", fc="yellow", alpha=0.5))

    plt.title('Dinamika Performa Model Selama Pelatihan (Epoch 1-10)', fontsize=14)
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Skor Korelasi Spearman', fontsize=12)
    plt.xticks(range(1, len(df_full)+1)) # Paksa sumbu X bilangan bulat
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    plt.tight_layout()
    plt.savefig("grafik_full_training.png", dpi=300)
    plt.show()
    print("Grafik disimpan: grafik_full_training.png")
```
